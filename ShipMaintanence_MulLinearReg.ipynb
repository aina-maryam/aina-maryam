```python
from sklearn.model_selection import train_test_split
```

```python
df['Total Factory Call Cost $'] = df['Total Factory Call Cost $'].fillna(df['Total Factory Call Cost $'].mean())
df['Total Commercial Call Cost $'] = df['Total Commercial Call Cost $'].fillna(df['Total Factory Call Cost $'].mean())
mean_value = df['Material Factory Cost $'].mean() 
# Replace 0 values with the mean value 
df['Material Factory Cost $'] = df['Material Factory Cost $'].replace(0, mean_value)
```

```python
print(df['Total Commercial Call Cost $'].isna().sum())
```

```python
X = df [['Total Factory Call Cost $','Labor Hours','Total Hours']]
y = df['Total Commercial Call Cost $']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.3, random_state = 42)
```

```python
X_train
```

```python
X_test
```

```python
# training the model
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
lm.fit(X_train, y_train)
```

```python
lm.coef_
```

```python
# predictions
y_pred = lm.predict(X_test)
```

```python
y_pred_df = pd.DataFrame(y_pred,columns=['Predicted Values'])
print(y_pred_df)
```

```python
y_test
```

```python
sns.scatterplot(x = y_pred, y = y_test, alpha = 0.3)
plt.xlabel("Predictions")
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
import math
```

```python
print('Mean Squared Error: ', mean_squared_error(y_test,y_pred))
print('Mean Absolute Error: ', mean_absolute_error(y_test,y_pred))
print('Root Mean Square Error: ', math.sqrt(mean_squared_error(y_test, y_pred)))
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_pred)
print('R-squared:', r2)
```
