# cust_sat.csv
- cust id
- products (books, electronics, clothes)
- age (18-70)
- score (1-5)
- gender (M and F)

```python
# Import libraries
import pandas as pd
import numpy as np

# Generate the data
np.random.seed(42)
cust_id = np.arange(1,501)
product = np.random.choice(['Books', 'Electronics', 'Clothes'], size=500)
age = np.random.randint(18, 71, 500)
score = np.random.randint(1, 6, 500)
gender = np.random.choice(['Male','Female'], 500)
```

```python
# Create data

df = pd.DataFrame({'cust_id': cust_id,
                   'product': product,
                   'age': age,
                   'score': score,
                   'gender': gender})

df.to_csv('customer.csv', index=False)
```

```python
df = pd.read_csv('customer.csv')
df.head()
```

# 1. Descriptive Analysis

```python
df.describe(include='all')
```

# 2. T-Test

 - H0 - There is no mean significance difference in the satisfaction score of males and females.
 - H1 - There is a mean significance difference in the satisfaction score of males and females.


```python
# Independent T-Tests
from scipy.stats import ttest_ind

# split the group
male = df[df['gender'] == 'Male']['score']
female = df[df['gender'] == 'Female']['score']

t_stats, p_val = ttest_ind(male, female)

print('T-Statistics: ', t_stats)
print('P-Value: ', p_val)
```

# Anova Test
- sample number is 3 or more than 3
- mean difference between the groups
- mean difference within the groups
- ratio of both is called f statistics value
- one way, two way, MANOVA (multi)

- H0 - Mean satisfaction score of all products are same.
- H1 - Mean satisfaction score of all products are different.

```python
from scipy.stats import f_oneway

# split the products
books = df[df['product'] == 'Books']['score']
electronics = df[df['product'] == 'Electronics']['score']
clothes = df[df['product'] == 'Clothes']['score']

anova = f_oneway(books, electronics, clothes)
print(f'ANNOVA Results - F-Statistics: {anova.statistic}, P-Value: {anova.pvalue}')
```

# Chi Square
- Analysis of the groups that are categorical.
- H0 - The is no association between gender and products.
- H1 - The is association between gender and products.

```python
from scipy.stats import chi2_contingency

table = pd.crosstab(df['product'], df['gender'])

chi2, p, dof, expected = chi2_contingency(table)

print(f'Chi-Square Results - Chi-Square: {chi2}, P-Value: {p}')
```
